```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
for i in range(1,9):
    df = pd.read_excel(rf'CSW8-F21-reflections/anon_Week-{i}-Reflection.xlsx', index_col=0)
    print(len(df))
```

```python
# remove stopwords
# calculate n-gramm freq abive all n = 1,2,3, 4?
# select top n = 10 n gramms
```

```python
only_text_answer = pd.DataFrame(df[df.columns[1]])
only_text_answer.columns = ["initial"]
```

```python

def replace_all(text):
    text = text.lower()
    
    for simb in (".", ",", "!", "?", ";", ":", "(", ")"):
        text = text.replace(simb, " ")
    text = " " + text + " "
    for word in stopwords.words('english'):
        word = " " + word + " "
        text = text.replace(word, " ")
    
    text = text.replace("   ", " ")
    text = text.replace("  ", " ")
    return text.strip()
```

```python
sample = only_text_answer["initial"][0]
print(sample)
print(replace_all(sample))
```

```python
only_text_answer[only_text_answer["initial"] == None] = ""
only_text_answer[only_text_answer["initial"].isna()] = ""
only_text_answer["no_stop"] = only_text_answer["initial"].apply(replace_all)
```

```python
only_text_answer["list"] = only_text_answer["no_stop"].apply(lambda x: x.split())
```

```python
only_text_answer["list"]
```

```python
freq = dict()
for size in range(2,4):
    for line in only_text_answer["list"]:
        for k in range(len(line)-size):
            if len(line) >= size:
                key = tuple(line[k:k+size])
                freq[key] = freq.get(key, 0) + 1
```

```python
freq = [(key, val) for key, val in freq.items()]
sorted(freq, key=lambda x: x[1], reverse=True)
```
